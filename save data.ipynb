```python
import os
import json
from tqdm import tqdm
import re
from unidecode import unidecode
```

```python
exclude_user = open("exclude_user.txt", "r").read().split("\n")
```

```python
exclude_tweet = [
    "MOM ",
    "SPORTS ",
    "HOBBIES ",
    "AUTO ",
    "FASHION "
    "MENS STYLE ",
    "TRAVEL ",
    "BEAUTY ",
    "TECHNO ",
    "FINANCE ",
    "FOOD ",
    "GADGET ",
    "blackpink ",
    "saya menjual",
    "Saya menjual",
    "Yuk Sobee",
    "Yuks Monggo",
    "Yuk",
    "READY STOCK",
]
```

```python
def cleaning_tweet(tweet):
    clean_tweet = re.sub("@+[A-z]+", '', str(tweet))

    # Remove url
    clean_tweet = re.sub(r"https:\/\/.*[\r\n]*", '', clean_tweet)

    # Remove #hashtag
    clean_tweet = re.sub(r"([@#][\w_-]+)", '', clean_tweet)

    # Remove Emoji
    emoji_pattern = re.compile("["
    u"\U0001F600-\U0001F64F"  # emoticons
    u"\U0001F300-\U0001F5FF"  # symbols & pictographs
    u"\U0001F680-\U0001F6FF"  # transport & map symbols
    u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                       "]+", flags=re.UNICODE)
    clean_tweet = emoji_pattern.sub(r'', clean_tweet)
    
    clean_tweet = clean_tweet.replace("\\n", " ").replace("b'","")
    clean_tweet = clean_tweet.replace("&amp;", "").replace("   "," ").replace("  ", " ")
    clean_tweet = clean_tweet.replace(" , ",", ").replace('"', "").replace("\n", "").strip()
    clean_tweet = unidecode(clean_tweet)
    
    return clean_tweet
```

```python
def check_tweet(tweet):
    for et in exclude_tweet:
        tweet_split = tweet.split(" ")[0]
        if tweet_split == "Jual":
            tweet = "discard"
            break
        elif len(tweet) < 5:
            tweet = "discard"
            break
        elif et in tweet:
            tweet = "discard"
            break
    
    return tweet
```

```python
all_data = ["discard"]
id_ = 0
# user_name = {}
fsave = open("data/all_corpus.jsonl", "w")
for name in os.listdir("final/"):
    if ".txt" in name:
        raw = open("final/{}".format(name), "r")
#         print(name.replace(".txt", ""))
        data = raw.readlines()
        raw.close()
        for i in tqdm(range(len(data)), desc=name.replace(".txt", "")):
            user = json.loads(data[i])["user"]["screen_name"].lower()
            tweet = json.loads(data[i])["full_text"]
#             try:
#                 user_name[user] +=1
#             except:
#                 user_name[user] = 0
#                 user_name[user] +=1
            
            if user in exclude_user:
                pass
            else:
                tweet = cleaning_tweet(tweet)
                tweet = check_tweet(tweet)
                if tweet in all_data:
                    pass
                else:
                    id_ += 1
                    res = {
                        "id_": id_,
                        "tweet": tweet,
                        "user": user
                    }
                    fsave.write(json.dumps(res)+"\n")
                    all_data.append(tweet)
                    
fsave.close()
```
